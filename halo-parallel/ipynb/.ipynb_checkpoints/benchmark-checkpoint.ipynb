## Benchmark Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# path = "../halo_sim_parallel_benchmark.record"

path = "../gxx/benchmark-banhcall/euler-parallel-benchmark-openmp-bahcall.record"
```

```python
df = pd.read_csv(path, delimiter=', ')
df
```

```python
def barcharts(n):
    
    
    dfsub = df.loc[ (df['Ntop'] == n) ]

    dfsub[ dfsub['Size'] == 1000 ].plot.bar(x='OMPThread',y='Time')
    plt.title('Steps 1000')

    dfsub[ dfsub['Size'] == 10000 ].plot.bar(x='OMPThread',y='Time')
    plt.title('Steps 10000')

    dfsub[ dfsub['Size'] == 20000 ].plot.bar(x='OMPThread',y='Time')
    plt.title('Steps 20000')

    dfsub[ dfsub['Size'] == 30000 ].plot.bar(x='OMPThread',y='Time')
    plt.title('Steps 30000')

    dfsub[ dfsub['Size'] == 40000 ].plot.bar(x='OMPThread',y='Time')
    plt.title('Steps 40000')


    plt.show()
```

### df100

```python
df100 = df.loc[ (df['Ntop'] == 100) ]
df100
```
