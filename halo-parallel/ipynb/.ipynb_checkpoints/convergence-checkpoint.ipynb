```python
totlen1 = len(data1)
length1 = int(len(data1[0])/2)

totlen2 = len(data2)
length2 = int(len(data2[0])/2)
```

```python
x1 = data1[0][:length1+1]
x2 = data2[0][:length2+1]
```

```python
def plotdata(x,data,length,style,lab):
    
    plt.plot(x, data[-3][:length+1]/2+0.5, lines[style]+'r', markersize =5, label='Forward ' + lab)
    plt.plot(x[1:length+1], data[-3][length:2*length][::-1]/2+0.5, lines[style]+'k', markersize =5, label='Backward ' + lab)
```

```python
plotdata(x1,data1,length1,0,'avg_alpha 0.5')
plotdata(x2,data2,length2,6,'No Avg')

plt.title('Average Method Test ($R=0.1$; $\mu=1.0$)')

plt.ylim([0,1])
plt.legend(loc=8, ncol=2)

plt.show()
```
