### Vacuum Oscillations

```python
s2thetav = 0.916515138991;
omegav = 1;

def ProbTh(x):
    
    return 1 - ( s2thetav**2 )* np.sin( omegav * x / 2 )**2
```

```python
length = 100;
x = np.arange(0,10,5/length)
# print(x)
ptheory = ProbTh(x)
```

```python
plt.plot( x[:length],ptheory[:length] , lines[0]+"r", label='Vacuum Forward' )
plt.plot( x[0:length][::-1], ptheory[length:] , lines[0]+"k", label='Vacuum Back')
plt.title('Vacuum with Turn Over')
plt.show()
```
